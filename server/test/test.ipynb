```python
import requests
import json

# Datei mit den weather codes
weathercodes = json.load(open("weathercodes.json", "r"))

# Koordinaten (z.B. Berlin)
lat, lon = 52.52, 13.41

response = requests.get(f"https://api.open-meteo.com/v1/forecast?latitude={lat}&longitude={lon}&current=temperature_2m,windspeed_10m,weathercode")
data = response.json()
current = data["current"]
print(f"Temperatur: {current['temperature_2m']}°C | Wind: {current['windspeed_10m']} km/h | Wetter: {weathercodes[str(current["weathercode"])]}")
```

```python
import requests
import json

weathercodes = json.load(open("weathercodes.json", "r", encoding="utf-8"))

lat, lon = 49.4621020310017, 8.47714801139806

input = input("Type TEMPERATURE for the temperature of Mannheim Steubenstrasse 40")

if input == "TEMPERATURE":
    temp = requests.get(f"https://api.open-meteo.com/v1/forecast?latitude={lat}&longitude={lon}&current=temperature_2m").json()["current"]["temperature_2m"]
    if temp > 20:
        print (f"Das wetter heute beträgt schöne {temp} Grad Celsius")
    elif temp < 20:
        print (f"Das wetter heute beträgt leider recht kalte {temp} Grad Celsius :(")
elif input == "Weather":

    weather = requests.get(f"https://api.open-meteo.com/v1/forecast?latitude={lat}&longitude={lon}&current=temperature_2m,windspeed_10m,weathercode")
    data = weather.json()
    current = data["current"]
    weathercode = current["weathercode"]

    print(weathercodes[str(weathercode)])
```
